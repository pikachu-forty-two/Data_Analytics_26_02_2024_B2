### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()
print(males)
print(females)
```

```python
# sum the instances of males and females
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

# put them into a list called proportions
proportions = [males, females]

# Create a pie chart
plt.pie(
    # using proportions
    proportions,

    # with the labels being officer names
    labels = ['Males', 'Females'],

    # with no shadows
    shadow = False,

    # with colors
    colors = ['blue','red'],

    # with one slide exploded out
    explode = (0.15 , 0),

    # with the start angle at 90%
    startangle = 90,

    # with the percent listed as a fraction
    autopct = '%1.1f%%'
    )

# View the plot drop above
plt.axis('equal')

# Set labels
plt.title("Sex Proportion")

# View the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Generate sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Define a function to update the plot with a dynamic vertical line and x-axis value
def update_plot(highlight_x):
    # Clear the previous plot
    plt.clf()

    # Plot histogram
    plt.hist(data, bins=30, color='skyblue', edgecolor='black')

    # Highlight specific x-value
    plt.axvline(x=highlight_x, color='red', linestyle='--', linewidth=2)

    # Add text annotation for the highlighted x-value
    plt.text(highlight_x, plt.ylim()[1]*0.9, f'X: {highlight_x:.2f}', color='red', fontsize=10, ha='center')

    # Set labels and title
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram with Highlighted Data Point')

    # Show the updated plot
    plt.show()

# Create an interactive slider to control the position of the line
highlight_slider = widgets.FloatSlider(value=0, min=min(data), max=max(data), step=0.1, description='Highlight X:')
widgets.interactive(update_plot, highlight_x=highlight_slider)
```
